```python
def get_plot_coords(wall_coords):
    alphabeth = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    y_idx = int(wall_coords.split("-")[0])-1
    x_idx = alphabeth.find(wall_coords.split("-")[1])
    h_cumulative_hole_distance, v_cumulative_hole_distance = get_coordinates()
    return float(h_cumulative_hole_distance[x_idx]),float(v_cumulative_hole_distance[y_idx])

def get_plot_color(hold_color):
    if hold_color == "blue":
        plot_color = "#4169E1"
    elif hold_color == "mint":
        plot_color = "#3EB489"
    elif hold_color == "neon":
        plot_color = "#FFFF33"
    else:
        plot_color = hold_color
    return plot_color

def get_plot_size(hold_size):
    if hold_size == "L":
        plot_size = 200
    elif hold_size == "M":
        plot_size = 130
    elif hold_size == "S":
        plot_size = 60
    elif hold_size == "bolt":
        plot_size = 2
    else:
        print("Could not identify hold size.")
    return plot_size

def get_coordinates():
    h_hole_distance = [15]+[20]*11
    h_cumulative_hole_distance = [np.sum(h_hole_distance[:i+1]) for i in range(len(h_hole_distance))]
    v_hole_distance = [5]+[20]*14
    v_cumulative_hole_distance = [np.sum(v_hole_distance[:i+1]) for i in range(len(v_hole_distance))]
    return h_cumulative_hole_distance, v_cumulative_hole_distance

def create_home_wall(griffe):
    # create dictionary with all holds & their properties
    h_cumulative_hole_distance, v_cumulative_hole_distance = get_coordinates()
    home_wall = {}
    for letter in "ABCDEFGHIJKL":
        for number in range(1,16):
            hold_coords = f"{number}{letter}"
            old_hold_coords = f"{number}-{letter}"
            plot_coords = get_plot_coords(old_hold_coords)

            if hold_coords in griffe:
                hold_mounted = True
                hold_color = griffe[hold_coords][0]
                hold_size = griffe[hold_coords][1]
            else:
                hold_mounted = False
                hold_color = "grey"
                hold_size = "bolt"

            plot_color = get_plot_color(hold_color)
            plot_size = get_plot_size(hold_size)
            home_wall[hold_coords] = {
                        "hold_mounted": hold_mounted,
                        "hold_color": hold_color,
                        "hold_size": hold_size,
                        "plot_coords": plot_coords,
                        "plot_color": plot_color,
                        "plot_size": plot_size,
            }
    return home_wall
```

```python
def load_hold_setup(version="newest"):
    saved_hold_setups = os.listdir("saved_hold_setups")
    if version == "newest":
        version = sorted(saved_hold_setups)[-1]
    with open(os.path.join('saved_hold_setups',version), 'r') as f:
        # write the dictionary to the file in JSON format
        home_wall = json.load(f)
    return home_wall
```
